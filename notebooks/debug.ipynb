```python
import os
import json
import pandas as pd
import logging
from datetime import datetime

output_dir = "/Users/vince/Salk/PaperGeneration/data/output"
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

results = []  # List to store results for analysis

for dir_name in os.listdir(output_dir):
    dir_path = os.path.join(output_dir, dir_name)
    if not os.path.isdir(dir_path):
        continue

    logger.info(f"Processing directory: {dir_name}")

    final_jsons = [f for f in os.listdir(dir_path) if "_final_" in f and f.endswith(".json")]
    papers_jsons = [f for f in os.listdir(dir_path) if "_papers_" in f and f.endswith(".json")]

    if not final_jsons or not papers_jsons:
        logger.warning(f"No _final_ or _papers_ JSON files found in {dir_name}. Skipping.")
        continue

    # Find the latest final JSON
    latest_final_json = None
    latest_final_time = None

    for f in final_jsons:
        try:
            dt = datetime.strptime(f.split('_')[-2] + ' ' + f.split('_')[-1].split('.json')[0], '%Y%m%d %H%M%S')
            if latest_final_time is None or dt > latest_final_time:
                latest_final_time = dt
                latest_final_json = f
        except ValueError as e:
            logger.error(f"Error parsing filename {f}: {e}")
            continue

    # Find the latest papers JSON
    latest_papers_json = None
    latest_papers_time = None

    for f in papers_jsons:
        try:
            dt = datetime.strptime(f.split('_')[-2] + ' ' + f.split('_')[-1].split('.json')[0], '%Y%m%d %H%M%S')
            if latest_papers_time is None or dt > latest_papers_time:
                latest_papers_time = dt
                latest_papers_json = f
        except ValueError as e:
            logger.error(f"Error parsing filename {f}: {e}")
            continue

    if latest_final_json and latest_papers_json:
        final_json_path = os.path.join(dir_path, latest_final_json)
        papers_json_path = os.path.join(dir_path, latest_papers_json)

        try:
            with open(final_json_path, 'r') as f:
                final_data = json.load(f)
            final_char_count = os.path.getsize(final_json_path) # More efficient than len(json.dumps())
            json_valid = True
        except json.JSONDecodeError as e:
            logger.error(f"Invalid JSON in {latest_final_json}: {e}")
            final_char_count = os.path.getsize(final_json_path)
            json_valid = False
        except FileNotFoundError:
            logger.error(f"File not found {final_json_path}")
            continue

        try:
            with open(papers_json_path, 'r') as f:
                papers_data = json.load(f)
            papers_char_count = os.path.getsize(papers_json_path)
        except json.JSONDecodeError as e:
            logger.error(f"Invalid JSON in {latest_papers_json}: {e}")
            papers_char_count = os.path.getsize(papers_json_path)
        except FileNotFoundError:
            logger.error(f"File not found {papers_json_path}")
            continue

        results.append({
            "directory": dir_name,
            "final_json_file": latest_final_json,
            "final_char_count": final_char_count,
            "papers_json_file": latest_papers_json,
            "papers_char_count": papers_char_count,
            "json_valid": json_valid,
            "final_json_datetime": latest_final_time,
            "papers_json_datetime": latest_papers_time
        })
    else:
        logger.warning(f"Could not determine latest JSON files in {dir_name}")

# Create a Pandas DataFrame for easy analysis
df_results = pd.DataFrame(results)
print(df_results)
# Now you can analyze df_results to find correlations
# For example:
print("\nValue Counts for json_valid")
print(df_results["json_valid"].value_counts())

print("\nCorrelation between final_char_count and json_valid")
print(df_results[["final_char_count", "json_valid"]].corr())

output_csv_path = os.path.join(output_dir, "analysis_results.csv")
df_results.to_csv(output_csv_path, index=False)
logger.info(f"Analysis results saved to {output_csv_path}")
```

```python
df_results['total_count'] = df_results['final_char_count'] + df_results['papers_char_count']
```

```python

# Convert json_valid to numerical for easier filtering
df_results['json_valid_numeric'] = df_results['json_valid'].map({True: 1, False: 0})

# Separate data for valid and invalid JSONs
valid_jsons = df_results[df_results['json_valid']]
invalid_jsons = df_results[~df_results['json_valid']]  # ~ is the negation operator

# Summary statistics for valid JSONs
logger.info("Summary Statistics for Valid JSONs:")
if not valid_jsons.empty:
    print(valid_jsons[['final_char_count', 'papers_char_count']].describe())
else:
    logger.info("No valid JSONs found.")

# Summary statistics for invalid JSONs
logger.info("\nSummary Statistics for Invalid JSONs:")
if not invalid_jsons.empty:
    print(invalid_jsons[['final_char_count', 'papers_char_count']].describe())
else:
    logger.info("No invalid JSONs found.")


# More concise way using groupby
summary_stats = df_results.groupby('json_valid')[['final_char_count', 'papers_char_count']].describe()
logger.info("\nSummary Statistics using groupby:")
display(summary_stats)

# If you only need min, max, and mean:
min_max_mean = df_results.groupby('json_valid')[['final_char_count', 'papers_char_count']].agg(['min', 'max', 'mean'])
logger.info("\nMin, Max, Mean using groupby:")
display(min_max_mean)


# If you only need min, max, and mean:
min_max_mean = df_results.groupby('json_valid')[['final_char_count', 'papers_char_count', 'total_count']].agg(['min', 'max', 'mean'])
logger.info("\nMin, Max, Mean using groupby:")
display(min_max_mean)
```
