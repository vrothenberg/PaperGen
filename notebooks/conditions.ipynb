```python
df = pd.read_csv("condition_revised.csv").drop_duplicates().fillna("")
df
```

```python
df['Category'].value_counts()
```

```python
# Strip leading and trailing spaces 

df['Condition'] = df['Revised Condition'].str.strip()
df['Category'] = df['Category'].str.strip()
df['Alternative'] = df['Alternative Name'].str.strip()
 
new_df = df[['Condition', 'Alternative', 'Category']].drop_duplicates().fillna("")
new_df
```

```python
counts = new_df['Category'].value_counts()
counts
```

```python
# Show only the rows where the category is less than 5 

filtered_df = new_df[new_df['Category'].isin(counts[counts < 4].index)]
filtered_df
```
