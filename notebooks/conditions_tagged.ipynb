```python
import pandas as pd
from collections import Counter

# 1. Load the CSV
df = pd.read_csv("condition_tagged.csv")

# 2. Extract and Flatten the Tags
all_tags = []
for tags_str in df["Tags"]:
    tags_list = tags_str.split("; ")  # Split the string by "; "
    all_tags.extend(tags_list)  # Add the individual tags to the all_tags list

# 3. Count Tag Occurrences
tag_counts = Counter(all_tags)

# Or, using Pandas:
# tag_counts = pd.Series(all_tags).value_counts()

# 4. Display the Results
print("Tag Counts:")
for tag, count in sorted(tag_counts.items()):
    print(f"{tag}: {count}")
```
